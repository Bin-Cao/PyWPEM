```python
import sys
import pandas as pd
Yourdir = '/Users/jacob/Documents/GitHub/'
sys.path.append(Yourdir)
from PyWPEM import WPEM
```

```python
intensity_csv = pd.read_csv(r'int.csv',header=None )
var = WPEM.BackgroundFit(intensity_csv,bac_split=7,lowAngleRange=905,bac_num=180,Model='XPS',noise = 0.3,bac_var_type='multivariate gaussian')
```

```python
AtomIdentifier = [['Cu','2p3/2',932.9],['Cu','2p1/2',952.5],]

# The file name of non-background data 
no_bac_intensity_file = "no_bac_intensity.csv" 
# The file name of raw/original data 
original_file = "int.csv"  
# The file name of background data 
bacground_file = "bac.csv"  



# Execute the model
WPEM.XPSfit(
    var, AtomIdentifier, no_bac_intensity_file, original_file, bacground_file, 
    bta = 0.90,iter_max = 100, asy_C = 0.,
    )
```
